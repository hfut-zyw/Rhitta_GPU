# sinx数据集

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 数据集介绍  
数据集构造：从sin(x)函数上取样300个点，并加入噪声  
数据集用途：1.拟合实验 2.当成气温的时间序列预测实验

### 数据采样

```python
N = 300 # 300个点
noise = 0.1 # 为了结果好看点，噪音取小一点
x = np.linspace(0, 6.28, N)
y = np.sin(x) + np.random.normal(0, noise, x.shape)
plt.figure(0)
plt.subplot(121)
plt.plot(x, np.sin(x), label='sin(x)')
plt.subplot(122)
plt.plot(x, y, label='sin(x)+noise')
```
